```python
# Seleccionar las columnas relevantes para el análisis
diagnosis_columns = ['OASISID', 'NORMCOG', 'MCIAMEM', 'MCIAPLUS', 'MCIAPLAN', 'DEMENTED']

# Agrupar por OASISID y calcular el valor máximo para cada diagnóstico (1 si alguna visita tiene el diagnóstico)
grouped_diagnosis_subset = diag_df[diagnosis_columns].groupby('OASISID').max()

# Crear una nueva columna para agrupar todas las condiciones de MCI en una sola categoría
grouped_diagnosis_subset['MCI_ALL'] = grouped_diagnosis_subset[['MCIAMEM', 'MCIAPLUS', 'MCIAPLAN']].max(axis=1)

# Crear una categoría "No Presenta" para los sujetos que no tienen diagnóstico
grouped_diagnosis_subset['NAN'] = (grouped_diagnosis_subset[['NORMCOG', 'DEMENTED', 'MCI_ALL']].sum(axis=1) == 0).astype(int)

# Agrupar las columnas en cuatro categorías: Cognición Normal, Todos los MCI, Demente, y No Presenta
summary = pd.DataFrame({
    'Cognición Normal': grouped_diagnosis_subset['NORMCOG'].sum(),
    'Todos los MCI': grouped_diagnosis_subset['MCI_ALL'].sum(),
    'Demented': grouped_diagnosis_subset['DEMENTED'].sum(),
    'NaN': grouped_diagnosis_subset['NAN'].sum()
}, index=['Número de Sujetos']).T.reset_index()

# Crear el gráfico de barras con seaborn
plt.figure(figsize=(16, 10))
sns.barplot(data=summary, x='index', y='Número de Sujetos', hue='index')

# Personalizar el gráfico
plt.title('Distribución de Diagnósticos Cognitivos por Categoría')
plt.xlabel('Condición Cognitiva')
plt.ylabel('Número de Sujetos')
plt.tight_layout()
plt.show()
```

```python
progression_columns = ['OASISID', 'days_to_visit', 'OASIS_session_label',  'NORMCOG', 'MCIAMEM', 'MCIAPLUS', 'MCIAPLAN', 'DEMENTED']

progression_data = diag_df[progression_columns].sort_values(by=['OASISID', 'days_to_visit'])

progression_data.fillna(0, inplace=True)

progression_data['progression'] = progression_data.apply(
    lambda row: 'MCI' if row[['MCIAMEM', 'MCIAPLUS', 'MCIAPLAN']].max() == 1 else (
        'Demented' if row['DEMENTED'] == 1 else 'Normal'
    ), axis=1
)
progression_changes = progression_data.groupby('OASISID').filter(
    lambda x: (x['progression'] == 'Normal').any() and ((x['progression'] == 'MCI').any() or (x['progression'] == 'Demented').any())
)
def check_progression(df):
    if (df['NORMCOG'].iloc[0] == 1) and (df['progression'] != 'Normal').any():
        return True
    return False

deteriorating_patients = progression_data.groupby('OASISID').filter(check_progression)

deterioration_count = deteriorating_patients['OASISID'].nunique()

deterioration_table = deteriorating_patients[['OASISID', 'OASIS_session_label', 'days_to_visit', 'progression']].drop_duplicates()

```

# Visualizing images

```python
import nibabel as nib
import matplotlib.pyplot as plt
import json
import os

def load_nifti_image(file_path):
    """Load a NIfTI (.nii or .nii.gz) file and return the image data."""
    img = nib.load(file_path)
    data = img.get_fdata()
    return data

def plot_slice(image_data, slice_index, title=""):
    """Plot a single slice of the 3D MRI image data."""
    plt.figure(figsize=(8, 8))
    if len(image_data.shape) == 3:  # Anatomical data
        plt.imshow(image_data[:, :, slice_index], cmap="gray")
    elif len(image_data.shape) == 4:  # Functional data, taking the first time point
        plt.imshow(image_data[:, :, slice_index, 0], cmap="gray")
    plt.title(title)
    plt.axis('off')
    plt.show()

def load_json_metadata(file_path):
    """Load a .json file and return the parsed data."""
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            metadata = json.load(f)
        return metadata
    except (UnicodeDecodeError, json.JSONDecodeError):
        print(f"Skipping non-JSON or improperly encoded file: {file_path}")
        return None

def visualize_mri_files(directory):
    """Visualize the MRI files and metadata from the given directory."""
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.startswith("._"):
                continue  # Skip hidden/system files
            if file.endswith(".nii.gz"):
                file_path = os.path.join(root, file)
                modality_type = "Anatomical" if "anat" in root else "Functional"
                print(f"Loading {modality_type} MRI image from: {file_path}")
                image_data = load_nifti_image(file_path)
                # Display the middle slice
                title = f"{modality_type} MRI - {os.path.basename(file_path)}"
                plot_slice(image_data, image_data.shape[2] // 2, title)
            elif file.endswith(".json"):
                file_path = os.path.join(root, file)
                print(f"Loading metadata from: {file_path}")
                metadata = load_json_metadata(file_path)
                if metadata is not None:
                    print(json.dumps(metadata, indent=4))


```
